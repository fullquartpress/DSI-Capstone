```python
def title_scraper(max_pages):
    page = 1
    while page < max_pages:
        path = '//*[@id="searchTop"]/div/div/section[3]/div[1]/div/a' 
        
             #'//*[@id="searchTop"]/div/div/section[3]/div[2]/div/a'
        
        browser.find_element_by_xpath(path)

    titles = browser.find_elements_by_class_name("search-result__heading article-preview__headline click-utag ng-binding")
    for title in titles:
        print(title.text)
```
